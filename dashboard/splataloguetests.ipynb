```python
frequency_ranges = [(680.35, 682.34), (681.29, 683.15), (688.19, 690.06)]  # Example range pairs

all_lines = []
#loop through the freq pairs
for start_freq, end_freq in frequency_ranges:
    #Splat query
    lines = Splatalogue.query_lines(start_freq*u.GHz, end_freq*u.GHz)
    lines_df = lines.to_pandas()
  
    #append df to list
    all_lines.append(lines_df)

#concatenate all df into big df
combined_df = pd.concat(all_lines, ignore_index=True)

combined_df
```
